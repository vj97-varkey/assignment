```python
import pandas as pd 
data = pd.read_csv(r"C:\Users\VARKEY JOSHY T\Desktop\data\fifa_data.csv")
data['Wage'] = data['Wage'].str.replace('€', '').str.replace('K', '').astype(float) *1000
highest_salary_player = data.loc[data['Wage'].idxmax()]
print(f"Highest paid player: {highest_salary_player['Name']} with salary of € {highest_salary_player['Wage']}")
```

```python
import pandas as pd 
data = pd.read_csv(r"C:\Users\VARKEY JOSHY T\Desktop\data\fifa_data.csv")
def height_to_inches(height_str):
    try:
        feet, inches = height_str.split("'")
        inches = inches.replace('"', '').strip()
        return int(feet) * 12 + int(inches)
    except Exception:  
        return None
tallest_player_data = data['Height'].apply(height_to_inches).idxmax()
tallest_player = data.loc[tallest_player_data]
print(f"The tallest player is: {tallest_player['Name']} with height of {tallest_player["Height"]}")
```

```python
top_club = data['Club'].value_counts().head(1)
print(top_club)
```

```python
preferred_foot = data['Preferred Foot'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(preferred_foot.index, preferred_foot.values)
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.title('Most Preferred Foot Among Players')
plt.show()
```
